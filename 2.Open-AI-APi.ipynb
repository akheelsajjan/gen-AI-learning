```python
from pydantic import BaseModel, Field
from typing import List
```

```python
class Explanation(BaseModel):
    title: str = Field(description="Short title of the explanation")
    summary: str = Field(description="Clear explanation under 120 words")
    key_points: List[str] = Field(description="3â€“5 bullet points")
```

```python
schema_description = """
Return ONLY valid JSON in the following format:

{
  "title": "string",
  "summary": "string",
  "key_points": ["string", "string", "string"]
}

Rules:
- summary must be under 120 words
- key_points must contain 3 to 5 items
- no extra keys
- no explanations outside JSON
"""
```

```python
final_prompt = f"""
SYSTEM ROLE:
You are an expert Generative AI teacher.

CONTEXT:
{schema_description}

TASK:
Explain how these concepts connect together.
"""
```

```python
from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": final_prompt}
    ],
    temperature=0.2
)
```

```python
raw_output = response.choices[0].message.content
raw_output
```

```python
import json

data = json.loads(raw_output)
data
```

```python
result = Explanation(**data)
```

```python
result.title
result.summary
result.key_points
```
